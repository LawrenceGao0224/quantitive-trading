```python
import os
import pandas_datareader as pdr
import pandas as pd
from dotenv import load_dotenv

load_dotenv()
AAPL = pdr.get_data_tiingo('AAPL',api_key = os.getenv('api_key'))
AAPL.tail()
```

```python
# 資料預處理，取近300天的資料

AAPL_adj=AAPL.iloc[:,5:9] 
AAPL_adj.columns = ['Close', 'High', 'Low', 'Open'] 
AAPL_adj = AAPL_adj.reset_index() #從多index變成單一index
AAPL_adj.index = AAPL_adj['date'] # 將index設為date
AAPL_adj = AAPL_adj.iloc[:,2:6] 
AAPL_adj_20 = pd.DataFrame(AAPL_adj)
AAPL_adj_20
```

```python
MA5 = talib.SMA(AAPL_adj_20.Close, timeperiod = 5)
MA5
```

```python
EVENINGSTAR = talib.CDLEVENINGSTAR(AAPL_adj_20.Open.values,AAPL_adj_20.High.values,AAPL_adj_20.Low.values,AAPL_adj_20.Close.values)
EVENINGSTAR[EVENINGSTAR!=0]
```
